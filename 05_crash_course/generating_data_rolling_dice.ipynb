# Rolling dice with Plotly

Rolling a fair six-side dice, there equally chances to get a number from 1 to 6. However, when rolling 2 dices, there numbers more likely to appear than others.
We will try to determine which numbers are more likely

```python
import plotly as plt

from random import randint
```

```python
# Creating Die Class

class Die:
    """A class representing a single die"""
    
    def __init__(self, num_sides=6):
        """Assume a six-sided die"""
        self.num_sides = num_sides
        
    def roll(self):
        """Return a random value between 1 and number of sides"""
        return randint(1, self.num_sides)
```

```python
# Create a D6

die = Die()

# Make some rolls, and store the result in a list
results = []
for roll_num in range(10000):
    result = die.roll()
    results.append(result)
    
# see the result in a list
print(results)
```

```python
frequencies = []
for value in range(1, die.num_sides+1):
    # count for each position , the amount of occurances
    frequency = results.count(value)
    # append in the list
    frequencies.append(frequency)

# position indicates the occurances
print(frequencies)
```

### plot a Histogram
generate a html file to see the result

```python
from plotly.graph_objs import Bar, Layout
from plotly import offline

# Visualize the results
x_values = list(range(1, die.num_sides+1))
data = [Bar(x=x_values, y=frequencies)]

x_axis_config = {"title":"Result"}
y_axis_config = {"title": "Frequency of Result"}
my_layout = Layout(
                title = "Result of rolling one D6 1000 times",
                xaxis=x_axis_config,
                yaxis=y_axis_config)
offline.plot({"data":data,
              "layout": my_layout},
              filename="d6.html")
```

```python
import matplotlib.pyplot as plt

# plot with plt a barchar with the frequencies of occurrences of a value
fig, ax = plt.subplots()
ax.bar(x_values, frequencies, align='center', alpha=0.5);
```

### Rolling 2 Dices

```python
# Create 2 D6 dices
die_1 = Die()
die_2 = Die()

# Roll both dices 1.000 times and store the result using list comprehension
results = [die_1.roll()+die_2.roll() for roll_num in range(1000)]
print(results)
```

```python
# calculate the frequencies of each apparance using list comprehension
# assuming that a random result will include, at least, 1 occurence of the min value and 1 of the max value

frequencies = [results.count(value) for value in range(min(results), max(results)+1)]
print(frequencies)
```

```python
# print result
x_values = list(range(min(results), max(results)+1))

fig, ax = plt.subplots(figsize = (10,5))
ax.bar(x_values, frequencies, align='center', alpha=0.5);
```

What if we roll 2 dices of different sizes (eg. 6-sides and 10-sides)

```python
die_1 = Die()
die_2 = Die(10)

# Roll both dices 1.000.000 times and store the result using list comprehension
results = [die_1.roll()+die_2.roll() for roll_num in range(1_000_000)]

# assuming that a random result will include, at least, 1 occurence of the min value and 1 of the max value
# count frequencies
frequencies = [results.count(value) for value in range(min(results), max(results)+1)]

# plot the bar graph
x_values = list(range(min(results), max(results)+1))

fig, ax = plt.subplots(figsize = (10,5))
ax.bar(x_values, frequencies, align='center', alpha=0.5);
```

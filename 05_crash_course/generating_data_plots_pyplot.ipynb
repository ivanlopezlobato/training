### Generating Data

```python
import matplotlib.pyplot as plt
```

```python
# list of values
squares = [1,4,9,16,25]
# create a figure and axes with the method subplots()
fig, ax = plt.subplots()
# plot the list
ax.plot(squares);

# this function opens Matplotlib viewer. Not required in the notebook to visualize the graph
plt.show()
```

Customize the plot

```python
squares = [1,4,9,16,25]
fig, ax = plt.subplots()

ax.plot(squares, linewidth = 3)

# Set chart title and label axes
ax.set_title("Square Numbers", fontsize = 14)
ax.set_xlabel("Values", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

# Set size of tick labels
ax.tick_params(axis = "both", labelsize = 14)
```

```python
# correcting the value in the x-axis
input_values = [1,2,3,4,5]
squares = [1,4,9,16,25]
fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth = 3);

# Set size of tick labels
ax.tick_params(axis = "both", labelsize = 14)
```

### Built-in styles

```python
plt.style.available
```

```python
# correcting the value in the x-axis
input_values = [1,2,3,4,5]
squares = [1,4,9,16,25]

# setting style
plt.style.use("seaborn")
fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth = 3);

# Set size of tick labels
ax.tick_params(axis = "both", labelsize = 14)
```

### Plotting and styling individual points

```python
plt.style.use("seaborn")
fig, ax = plt.subplots()
ax.scatter(2,4,s=200)
ax.tick_params(axis="both", labelsize=14);
```

plotting several points with scatter()

```python
x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]

plt.style.use("seaborn")
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=100);
```

calculating data automatically

```python
# first 1000 integers
x_values = range(1,1001)
y_values = [x**2 for x in x_values]

# plot scatter
plt.style.use("seaborn")
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10);

# set the range of each axis
ax.axis([0,1100,0,1100000]);
# axis method requires a list of 4 values, xmin, xmax, ymin, ymax
```

```python
# color can be defined
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c="red", s=10);
```

```python
# color can be also defined by RGB parameters [0-1]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=(0, 0.5, 0), s=10);
```

```python
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=(0.8, 0.2, 0.8), s=10);
```

Using colormap

```python
# pyplot has built-in colormaps

fig, ax = plt.subplots()
# colormap will take y_values and apply the colormap Blues
ax.scatter(x_values, y_values, c=y_values, cmap= plt.cm.Blues, s=10);
```

## Random walks

Use matplotlib to create a data visualization of a random walk

random walk: path wihout clear direction, determined by a seies of random decisions

for this we create a Class RandomWalk

```python
# plot randomwalk
import matplotlib.pyplot as plt

# random choice
from random import choice
```

```python
class RandomWalk:
    """A class to generate random walks"""
    
    def __init__(self, num_points=5000):
        """Initialize attributes of a walk"""
        self.num_points = num_points
        
        # All walk start at (0,0)
        self.x_values = [0]
        self.y_values = [0]
        
    def fill_walk(self):
        """Calculates all the points in the walk"""
        
        # Keep taking steps until the walk reaches the desired lenght
        while len(self.x_values) < self.num_points:
            
            # Decide which direction to go and how far to go in that direction
            x_direction = choice([-1,1])
            x_distance = choice([0,1,2,3,4])
            x_step = x_direction * x_distance
            
            y_direction = choice([-1,1])
            y_distance = choice([0,1,2,3,4])
            y_step = y_direction * y_distance
            
            # Reject moves that go nowhere
            if x_step == 0 & y_step == 0:
                continue
                
            # Calculate the new position
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step
            
            # Add the values
            self.x_values.append(x)
            self.y_values.append(y)
```

### Plotting the random walk

```python
# Make a random walk
rw = RandomWalk()
rw.fill_walk()
```

```python
# Plot the point in the walk
plt.style.use("classic")
fig, ax = plt.subplots()
ax.scatter(rw.x_values, rw.y_values, s =10);
```

```python
# Genereate several random walk with a while loop till stopped by user

while True:
    # Make random walk
    rw = RandomWalk()
    rw.fill_walk()
    
    #Plot the point of the walk
    plt.style.use("classic")
    fig, ax = plt.subplots()
    ax.scatter(rw.x_values, rw.y_values, s =10);
    plt.show()
    
    # Ask for another walk
    keep_running = input("Make anouther walk?(y/n): ")
    if keep_running == "n":
        break
```

### Styling the walk

- Coloring the points
- Plotting the starting and ending points
- Cleaning up the axes
- adding plot points
- altering the size to fill the screen

```python
# Make random walk with 50.000 points (instead of default 5000)
rw = RandomWalk(50_000)
rw.fill_walk()

#Plot the point of the walk
plt.style.use("classic")
fig, ax = plt.subplots(figsize=(16,9)) # increase the size of the figure
point_number = range(rw.num_points)    # create a range of each step in the walk
ax.scatter(rw.x_values,
           rw.y_values, 
           c = point_number,           # use pont_number as value of color
           cmap = plt.cm.Blues,        # use gradient Blues to visualize the walk
           edgecolors="none",          # remove the edge of the dots to see better the walk
           s =1)

# Emphasize the first and last points
ax.scatter(0, 0, c="green",  marker="o", s=100) # Every walk starts in point (0,0)
ax.scatter(rw.x_values[-1], rw.y_values[-1], c="red", marker="*", s=200);

# remove the axes
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
```

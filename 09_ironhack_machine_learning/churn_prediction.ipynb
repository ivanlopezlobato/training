## Data preprocessing

```python
train.head()
```

### standarize all numerical features

```python
from sklearn.preprocessing import StandardScaler
```

```python
# standarize
std = StandardScaler()
scaled_fitted = std.fit(train[num_cols])
```

```python
# Visualise
df["tenure"].hist(bins=20)
```

```python
pd.DataFrame(scaled_fitted.transform(train[num_cols]), columns=num_cols)["tenure"].hist(bins=20);
```

```python
train = train.reset_index(drop=True)
test = test.reset_index(drop=True)
```

```python
scaled_num_cols = pd.DataFrame(scaled_fitted.transform(train[num_cols]), columns=["scaled_"+col for col in num_cols])
```

```python
scaled_num_cols.head()
```

```python
train = train.merge(scaled_num_cols, left_index=True, right_index=True, how="inner")
```

### create dummy variables

```python
cat_cols
```

```python
train = pd.get_dummies(data = train,
                       columns = cat_cols,
                       drop_first=True)
```

```python
# function to run test set thorugh the same processing steps

def preprocess(df, trained_standarization_factors):
    scaled_cols = pd.DataFrame(trained_standarization_factors.transform(df[num_cols]), columns=["scaled_"+col for col in num_cols])
    df = df.merge(scaled_cols, left_index=True, right_index=True, how="inner")
    df = pd.get_dummies(data=df,
                        columns=cat_cols,
                        drop_first=True)
    return df
```

## Train model

```python
test = preprocess(test, scaled_fitted)
```

```python
test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# create X, y
train_y = train[target]
train_X = train.drop(columns=id_col + target)

test_y = test[target]
test_X = test.drop(columns=id_col + target)

# train
lr = LogisticRegression()
lr.fit(train_X, train_y)

# predict
pred = lr.predict(test_X)

# accuracy_score
accuracy_score(test_y, pred)
```
